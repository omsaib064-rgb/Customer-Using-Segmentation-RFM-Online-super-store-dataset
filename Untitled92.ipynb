```python
import pandas as pd
import os

file_path = r"C:\Users\dell\Downloads\Sample - Superstore.csv"

# Check file exists
if not os.path.exists(file_path):
    raise FileNotFoundError("File not found. Check path.")

# Load data
df = pd.read_csv(file_path, encoding="latin1")

# Preview data
print("First 5 Rows:\n", df.head())

# Shape
print("Shape:", df.shape)

# Column names
print("Columns:", df.columns)

# Data types and nulls
print("\nInfo:\n")
df.info()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors='coerce')

df.info()
```

```python
df = df.drop(columns=['Row ID'])
```

```python
4. Using LIST and TUPLE
```

```python
# List
columns_list = list(df.columns)
print(columns_list)

# Tuple
columns_tuple = tuple(df.columns)
print(columns_tuple)
```

```python
CELL 5: GROUP BY (CATEGORY)
```

```python
sales_category = df.groupby('Category')['Sales'].sum()
print(sales_category)
```

```python
2. Profit by Region
```

```python
profit_region = df.groupby('Region')['Profit'].sum()
print(profit_region)
```

```python
3. Sales by Segment
```

```python
sales_segment = df.groupby('Segment')['Sales'].sum()
print(sales_segment)
```

```python
4. Top 10 Products
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
print(top_products)
```

```python
6. VISUALIZATION (MATPLOTLIB + SEABORN)
```

```python
1. Sales by Category (Bar Chart)
```

```python
sales_category.plot(kind='bar', color='blue')
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

```python
2. Profit by Region (Bar Chart)
```

```python
sns.barplot(x=profit_region.index, y=profit_region.values)
plt.title("Profit by Region")
plt.show()
```

```python
3. Sales by Segment (Pie Chart)
```

```python
sales_segment.plot(kind='pie', autopct='%1.1f%%')
plt.title("Sales by Segment")
plt.show()
```

```python
4. Top Products (Horizontal Bar)
```

```python
top_products.plot(kind='barh', color='green')
plt.title("Top 10 Products by Sales")
plt.show()
```

```python
5. Discount vs Profit (Scatter Plot)
```

```python
sns.scatterplot(x='Discount', y='Profit', data=df)
plt.title("Discount vs Profit")
plt.show()
```

```python
4. HISTOGRAM (Sales Distribution)
```

```python
sns.histplot(df['Sales'], bins=20, kde=True)
plt.title("Sales Distribution")
plt.show()
```

```python
advanced charts: Boxplot and Heatmap.
```

```python
1. BOXPLOT (Sales Distribution by Category)
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Category', y='Sales', data=df)
plt.title("Sales Distribution by Category")
plt.show()
```

```python
2. BOXPLOT (Profit by Region)
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Region', y='Profit', data=df)
plt.title("Profit Distribution by Region")
plt.show()
```

```python
3. HEATMAP (Correlation Matrix)
```

```python
plt.figure(figsize=(8,5))

corr = df[['Sales', 'Profit', 'Quantity', 'Discount']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
4. COUNT PLOT (Extra Advanced)
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Segment', data=df)
plt.title("Count of Customers by Segment")
plt.show()
```

```python
5. PAIRPLOT (Advanced Analysis)
```

```python
sns.pairplot(df[['Sales', 'Profit', 'Quantity', 'Discount']])
plt.show()
```
